```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas.plotting import scatter_matrix
import os

names = ['MaxAreaLowIntensityBloom','MaxAreaModerateIntensityBloom','MaxAreaHighIntensityBloom','MaxCyanConcentration','Freq']
raw_reservoir_response = pd.read_csv(os.path.join('data_in', 'raw_708reservoir.csv'), usecols=names)
```

```python
raw_reservoir_response.hist()
plt.show()
```

```python
raw_reservoir_response.plot(kind='density', subplots=True, layout=(2,3), sharex=False)
plt.show()
```

```python
raw_reservoir_response.plot(kind='box', subplots=True, layout=(3,2), sharex=False, sharey=False)
plt.show()
```

```python
correlations = raw_reservoir_response.corr()
# plot correlation matrix
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,5,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```
